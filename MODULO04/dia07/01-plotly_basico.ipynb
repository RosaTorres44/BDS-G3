# PLOTLY EXPRESS


```python
import plotly.express as px

df = px.data.iris()
df.head(2)
```

```python
fig = px.scatter(df,x='sepal_width',y='sepal_length',color='species')

fig.show()
```

# GRAFICOS CON DATASET PENGUINS

```python
import pandas as pd
import seaborn as sns

penguins_df = sns.load_dataset('penguins')
penguins_df.head(2)
```

```python
fig = px.scatter(penguins_df,
                 x='bill_length_mm',
                 y='bill_depth_mm',
                 color='species',
                 title='Relación entre bill_length y bill_depth')
fig.show()
```
