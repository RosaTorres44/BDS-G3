```python
import plotly
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
penguins_df = sns.load_dataset('penguins')
penguins_df.head(2)
```

# COLORES Y ESTILOS

```python
fig = px.scatter(penguins_df,
                 x='flipper_length_mm',
                 y='bill_depth_mm',
                 color='species',
                 color_discrete_sequence=['red', 'green', 'blue'],
                 symbol='species',
                 title='Relación entre flipper_length_mm y bill_depth_mm')
fig.show()
```

# USAR TEMPLATES

```python
import plotly.io as pio
pio.templates
```

## para cambiar el tema por defecto

```python
pio.templates.default = 'plotly_dark'
```

# TITULOS  ETIQUETAS Y LEYENDAS

```python
fig = px.scatter(penguins_df,
                 x='flipper_length_mm',
                 y='body_mass_g',
                 color='species',
                 color_discrete_sequence=['red', 'green', 'blue'],
                 symbol='species',
                 title='Relación entre flipper_length_mm y body_mass_g')

fig.update_layout(
    title = 'Relación entre flipper_length_mm y body_mass_g',
    xaxis_title = 'longitud aletas mm',
    yaxis_title = 'masa coporal gr'
)
fig.show()
```

# ESCALAS COLORES Y TEMAS

```python
fig = px.density_heatmap(penguins_df,
                 x='bill_length_mm',
                 y='bill_depth_mm',
                 title='Relación entre bill_length_mm y bill_depth_mm')
fig.show()
```

agregamos escala de verdes

```python
fig = px.density_heatmap(penguins_df,
                 x='bill_length_mm',
                 y='bill_depth_mm',
                 title='Relación entre bill_length_mm y bill_depth_mm',
                 color_continuous_scale='Greens')
fig.show()
```

# PERSONALIZANDO COLORES

```python
fig = px.scatter(penguins_df,
                  x='bill_length_mm',
                  y='bill_depth_mm',
                  title='Bill Length vs Bill Depth',
                  color='species',
                  color_discrete_map={"Adelie": "purple", "Gentoo": "orange", "Chinstrap": "cyan"})
fig.show()
```

# ANOTACIONES Y TEXTOS EN LOS GRAFICOS

```python
fig = px.scatter(penguins_df,
                 x='flipper_length_mm',
                 y='body_mass_g',
                  title='Relación entre flipper_length_mm y body_mass_g',
                  color='species',
                  color_discrete_map={"Adelie": "purple", "Gentoo": "orange", "Chinstrap": "cyan"})

fig.add_annotation(x=210,y=4500, text='Ejemplo pinguino con aleta de 210', showarrow = True , arrowhead = 2)

fig.show()
```
