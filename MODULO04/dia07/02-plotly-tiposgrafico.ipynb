# GRAFICOS CON DATASET PENGUINS

```python
import pandas as pd
import seaborn as sns

penguins_df = sns.load_dataset('penguins')
penguins_df.head(2)
```

# grafico de dispersion

```python
import plotly.express as px
```

```python
fig = px.scatter(penguins_df,
                 x='flipper_length_mm',
                 y='body_mass_g',
                 color='species',
                 title='Relación entre bill_length y bill_depth')
fig.show()
```

# grafico de barras

```python
fig = px.bar(penguins_df,
                 x='species',
                 title='cantidad por especie')
fig.show()
```

# HISTOGRAMAS

```python
fig = px.histogram(penguins_df,
                 x='body_mass_g',
                 color='species',
                 title='distribucion de peso por especie')
fig.show()
```

# grafico de lineas o line plot

primero ordenamos

```python
penguins_df_sorted = penguins_df.sort_values(by=['island','body_mass_g'])
```

```python
fig = px.line(penguins_df,
                 x='species',
                 y='body_mass_g',
                 color='island',
                 title='variacion de peso por isla')
fig.show()
```

# grafico de pie

```python
fig = px.pie(penguins_df,
                 names='species' ,
                 title='proporcion de pinguinos por especie')
fig.show()
```
