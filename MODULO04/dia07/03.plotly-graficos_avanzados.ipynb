```python
import plotly
import plotly.express as px
plotly.__version__
```

carga dataset

```python
import pandas as pd
import seaborn as sns

penguins_df = sns.load_dataset('penguins')
penguins_df.head(2)
```

# GRADICO AREAS

```python
penguins_df_sorted = penguins_df.sort_values(by=['body_mass_g'])
```

```python
fig = px.area(penguins_df_sorted,
                 x='body_mass_g',
                 y='flipper_length_mm',
                 color='species',
                 title='longitud de aleta vs masa coporal')
fig.show()
```

# gradico de densidad

```python
fig = px.density_contour(penguins_df,
                 x='body_mass_g',
                 y='flipper_length_mm',
                 color='species',
                 title='densidad de masa coporal vs longitud de aleta')
fig.show()
```

# BOX PLOT

```python
fig = px.box(penguins_df,
                 x='species',
                 y='body_mass_g',
                 color='species',
                 title='Distribucion de pesos por especie')
fig.show()
```

# grafico de violin

```python
fig = px.violin( penguins_df,
                 x='species',
                 y='body_mass_g',
                 color='species',
                 box=True,
                 points='all',
                 title='Distribucion de pesos por especie')
fig.show()
```

#   MAPA DE CALOR

```python
import plotly.figure_factory as ff
import numpy as np


correlation = penguins_df.corr(numeric_only=True)
correlation
```

```python
fig = ff.create_annotated_heatmap(
                                  z=correlation.values,
                                  x=list(correlation.columns),
                                  y=list(correlation.index),
                                  colorscale='Viridis',
                                  annotation_text=np.round(correlation.values,2))
fig. show()
```

# GRAFICO 3D

```python
fig = px.scatter_3d(penguins_df,
                    x='bill_length_mm',
                    y='bill_depth_mm',
                    z='body_mass_g',
                    color= 'species',
                    title='Relación 3D entre bill_length bill_depth y body_mass')
fig. show()
```
