# tipos de datos en numpy

```python
arr = np.array([1,2,3,4])
arr.dtype
```

```python
arr_float = np.array([1,2,3,4], dtype='float64')
arr_float.dtype
```

# cast tipo de dato

```python
arr = np.array([1,2,3,4])
arr_bool = arr.astype(np.bool_)
arr_bool.dtype
```

```python
arr_bool
```

```python
arr = np.array([1,2,3,4])
arr_string = arr.astype(np.string_)
arr_string
```
