```python
from yahoo_fin import stock_info as si
import numpy as np
import matplotlib.pyplot as plt
```

jalamos data de bitcoin a dolares de un par de años

```python
data = si.get_data('BTC-USD', start_date = '2022-01-01', end_date='2024-12-31')
data
```

```python
dates = np.array(data.index, dtype='datetime64[D]')
dates
```

```python
closing_price = np.array(data['close'])
closing_price
```

```python
plt.figure(figsize=(12,6))
plt.title='Tendencia del precio de bitcoin en dolares'
plt.plot(dates,closing_price , label='Precio de Cierre', color='blue' )
plt.xlabel("fecha",fontsize=12)
plt.ylabel("precio cierre USD",fontsize=12)
plt.grid=True
plt.legend()
plt.show()

```
