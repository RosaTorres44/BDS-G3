# FILTROS Y CONDICIONES

```python
array = np.linspace(1,10,10,dtype='int8')
array
```



> filtros por condiciones
evalua cada scalar


```python
incide_condiciones = array > 5
incide_condiciones
```

```python
array[incide_condiciones]
```

otra forma de hacerlo es directo en el corchete

```python
array[array < 5]
```

```python
array[(array < 9) & (array > 5 )]
```

```python
array[(array < 9) & (array > 5 )] = 99
array
```
