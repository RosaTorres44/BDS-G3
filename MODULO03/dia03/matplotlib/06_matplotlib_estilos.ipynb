```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

# ESTILOS DE PYPLOT

```python
plt.style.available
```

```python
plt.style.use('dark_background')
```

```python
fig,axes = plt.subplots(figsize=(8,8))
axes.plot(x,x+1,'r--')
fig.show()
```

```python
fig,axes = plt.subplots(figsize=(8,8))
axes.plot(x,x+1,color='#15B01A',linestyle=':',marker='x',linewidth=5)
axes.plot(x,x+2,color='red',linestyle='--',marker='8',linewidth=8)
axes.plot(x,x+3,color='#1f77b4',linestyle='dashed',marker='x',linewidth=3)
axes.plot(x,x+4,color='#ff7f0e',linestyle='-',marker='o',linewidth=2)

fig.show()
```
