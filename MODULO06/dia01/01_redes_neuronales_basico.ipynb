# IMPORTAMOS LIBRERIAS BASE

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# IMPORTAMOS KERAS

```python
from keras.datasets import mnist
from keras import layers,models
from keras.utils import to_categorical
```

# CARGAMOS DATASET

```python
(train_data,train_labels),(test_data,test_labels) = mnist.load_data()
```

```python
train_data[0]
```

```python
plt.imshow(train_data[1])
plt.show()
```

```python
train_data.shape
```

# CREAMOS UN  MODELO DE REDES NEURONALES

```python
model = models.Sequential([
    layers.Input(shape=(28*28,)),  # 👈 Forma correcta
    layers.Dense(512, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
```

# CODIFICACIÓN Y ESCALAMIENTO DE DATOS

```python
x_train = train_data.reshape((60000,28*28))
x_train = x_train.astype('float32')/255
x_train
```

```python
x_test = test_data.reshape((10000,28*28))
x_test = x_test.astype('float32')/255
x_test
```

```python
y_train = to_categorical(train_labels)
y_train
```

```python
y_test = to_categorical(test_labels)
y_test
```

# ENTRENAMIENTO DE LA RED NEURONAL

```python
model.fit(x_train,y_train,epochs=5,batch_size=128)
```

# EVALUAMOS EL MODELO

```python
model.evaluate(x_test,y_test)
```

# REALIZAMOS UNA PREDICCION

```python
#seleccionamos una imagen de prueba
index = 2
image = x_test[index].reshape(28,28)
plt.imshow(image,cmap="gray")
```

```python
#realizamos la prediccion
prediction = model.predict(x_test[index].reshape(1,28*28))
predict_label = np.argmax(prediction)

#Mostramos la imagen y la prediccion
plt.imshow(image,cmap="gray")
plt.title(f"Prediccion: {predict_label}")
plt.show()
```
